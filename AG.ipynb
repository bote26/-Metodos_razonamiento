```python
from random import random
from typing import List 
import math
import random

def mapeo(a,b,L):
    paso = (b-a)/((2**L) -1)
    n = random.randint(0,2**L -1)
    x = a+n*paso
    return x

def generar_poblacion(n_individuos, tamaño_genes,a,b):
    poblacion = []
    for _ in range(n_individuos):
        individuo = mapeo(a, b, tamaño_genes)
        poblacion.append(individuo)
    return poblacion


def fitness(poblacion, función_objetivo):
    fitness_values = []
    for individuo in poblacion:
        valor_fitness = función_objetivo(individuo)
        fitness_values.append(valor_fitness)
    return fitness_values

def mutacion(poblacion, proba_mutacion):
    nueva_poblacion = []
    for individuo_str in poblacion:           
        individuo = list(individuo_str)       
        for i, c in enumerate(individuo):      
            if c in '01':                      
                if random.random() < proba_mutacion:
                    individuo[i] = '1' if c == '0' else '0'
        nueva_poblacion.append(''.join(individuo))
    return nueva_poblacion
    
def ruleta_padres(poblacion, fitness_values):
    total_fitness = sum(fitness_values)
    probas = [f / total_fitness for f in fitness_values]
    padres = random.choices(poblacion, weights=probas, k=len(poblacion))
    return padres
    
def cruce(padres, proba_cruce, punto_cruce):
    poblacion_nueva = []
    for i in range(0, len(padres), 2):  
        padre1 =padres[i]
        padre2 = padres[i+1]
        if random.random() < proba_cruce:
            pos_punto1 = padre1.find('.')
            pos_punto2 = padre2.find('.')
            padre1 = padre1.replace('.', '')
            padre2 = padre2.replace('.', '')
            hijo1 = padre1[:punto_cruce] + padre2[punto_cruce:]
            hijo2 = padre2[:punto_cruce] + padre1[punto_cruce:]
            hijo1 = hijo1[:pos_punto1] + '.' + hijo1[pos_punto1:]
            hijo2 = hijo2[:pos_punto2] + '.' + hijo2[pos_punto2:]
            poblacion_nueva.append(hijo1)
            poblacion_nueva.append(hijo2)
        else:
            hijo1, hijo2 = padre1, padre2
            poblacion_nueva.append(hijo1)
            poblacion_nueva.append(hijo2)
    return poblacion_nueva
        


def cambio_base(numero_completo, base, limite = 10):
    base_dict = {10:'A', 11:'B', 12:'C',13:'D',14:'E',15:'F'}
    lista_base = []
    num = math.trunc(numero_completo)
    num_decimal = numero_completo - num
    check_decimal = num_decimal
    while num > 0:
        modulo = num % base
        if modulo > 9:
            lista_base.append(base_dict[modulo])
        else:
            lista_base.append(str(modulo))
        num = num // base
   
    num = ''
    for i in lista_base[::-1]:
        num +=i
       
    lista_base_decimal = []
    count = 0
    while True:
        count +=1
        num_decimal *= base
       
        if num_decimal > 0:
            extraido = math.trunc(num_decimal)
            num_decimal -= extraido
            if extraido > 9:
                lista_base_decimal.append(base_dict[extraido])
            else:
                lista_base_decimal.append(str(extraido))
        elif num_decimal < 0:
            lista_base_decimal.append('0')
        else:
            break
        if count > limite:
            break
            
    num_decimal = ''
    for i in lista_base_decimal:
        num_decimal += i
   
    if check_decimal == 0:
        return num
    else:
        return f'{num}.{num_decimal}'

def funcion_objetivo(x):
    return x**2 + x +1
```

```python
N = 10
Tamaño_genoma = 5
G = 1
Proba_mutación = 0.1
Proba_cruce = 0.7
Punto_cruce = 2

poblacion = generar_poblacion(N, Tamaño_genoma, -5, 5)
print("Población inicial:")
print(poblacion)
fitness_values = fitness(poblacion, funcion_objetivo)
#print("Valores de fitness:")
#print(fitness_values)
elegidos = ruleta_padres(poblacion, fitness_values)
print("Padres seleccionados:")
print(elegidos)
elegidos_binario = []
for elegido in elegidos:
    elegido = abs(elegido)
    elegidos_binario.append(cambio_base(elegido,2))
print("Padres en binario:")
print(elegidos_binario)
nueva_pobla = cruce(elegidos_binario, Proba_cruce, Punto_cruce)
print("Población después del cruce:")
print(nueva_pobla)
pobla_mutada = mutacion(nueva_pobla, Proba_mutación)
print("Población después de la mutación:")
print(pobla_mutada)
pobla_final = []

for individuo in pobla_mutada:
    individuo_decimal = int(cambio_base(individuo,10))
    pobla_final.append(individuo_decimal)
print("Población final después de mutación:")
print(pobla_final)



```
